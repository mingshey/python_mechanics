## 오일러-라그랑주(Euler-Lagrange) 방정식

* Fowles, AM 10장
* 뉴턴의 표기법으로 시간에 대한 미분 $\dfrac{dy}{dt}$을 $\dot{y}$와 같이 쓰기로 한다.

라그랑지안이 시간의 직접적인 함수가 아닌 위치와 속도의 함수이고, 위치와 속도가 각각 시간의 함수일 때, 이것을
$$L = L(y,\ \dot{y};\ t)$$
라 쓰고, 라그랑지안의 적분(여기서는 시간 적분)을 "작용"(action)이라 한다. 라그랑지안과 작용은 각각 물체의 경로라고 하는 함수의 집합이 정의역인 함수, 즉 함수의 함수인데, 이것을 "범함수"(functional)라고도 한다. 라그랑지안과 작용의 정의역에는 경로를 나타내는 함수 뿐 아니라 경로의 시간에 대한 도함수도 포함된다.

작용이 최소가 되는 조건에 의해 물체의 경로는 경로의 변화(variation)에 대한 작용의 미분이 $0$인 경로를 따른다.

이 때 작용 또는 작용의 피적분함수인 라그랑지안은 범함수이므로 작용의 미분을 하기 위해서는 우선 작용의 정의역을 구성하는 경로($y(t)$) 및 경로의 시간 도함수($\dot y(t)$)에 대한 미분을 한 뒤 경로의 시간 미분을 곱하는 방식으로 연쇄 법칙을 적용해야 한다. 즉,
$$\begin{align} 0=\delta J &=\delta\int_{t_1}^{t_2} L dt\\
&=\int_{t_1}^{t_2} \delta L dt\\
&= \int_{t_1}^{t_2}\left(\frac{\partial L}{\partial y}\delta y + \frac{\partial L}{\partial \dot{y}}\delta\dot{y}\right)dt\\
&= \int_{t_1}^{t_2}{\color{blue}{\left(\frac{\partial L}{\partial y}\delta y + \frac{\partial L}{\partial \dot{y}}\frac{d\delta y}{dt}\right)}}dt. \end{align}$$



위 식의 마지막 단계에서
$$\delta \dot y = \frac{d\delta y}{dt}$$
를 사용했는데, 이를 설명하자면 다음과 같다:

실제 경로를 $y$, 변화된 경로를 $\bar y$라 하면 $\delta y = \bar y - y$ 라 할 수 있다.

$$\dot y = \frac{dy}{dt} = \lim_{dt\rightarrow 0}\frac{y(t+dt)-y(t)}{dt}$$에서 편의상 $\lim$를 생략하고
$$\dot y = \frac{dy}{dt} = \frac{y(t+dt)-y(t)}{dt}$$와 같이 쓰기로 하자. 그러면

$$
\begin{align}
\delta \dot y &= \delta\frac{d y}{dt} = \frac{d\bar y}{dt} - \frac{dy}{dt}\\
&= \frac{\bar y(t+dt)-\bar y(t)}{dt} - \frac{y(t+dt)-y(t)}{dt} \\
&= \frac{\bar y(t+dt)-y(t+dt)}{dt} - \frac{\bar y(t)-y(t)}{dt} \\
&= \frac{\delta y(t+dt)-\delta y(t)}{dt}\\
&= \frac{d\delta y}{dt}
\end{align}
$$
즉, $\dot y$의 변분은 $\delta y$의 시간 미분과 같고, 따라서 시간 미분과 변분의 순서를 바꾸어 적용할 수 있음을 알 수 있다.

$\color{blue}{파란색\ 괄호}$ 안의 두 번째 항은 부분적분에 의해

$$ \int_{t_1}^{t_2} \frac{\partial L}{\partial \dot{y}}\frac{d\delta y}{dt}dt = \left.\frac{\partial L}{\partial y}\delta y \right|_{t_1}^{t_2} - \int_{t_1}^{t_2} \frac{d}{dt}\frac{\partial L}{\partial \dot{y}}\delta ydt$$

인데, $\delta y$는 $t_0$, $t_1$에서 $0$이므로,

$$ \int_{t_1}^{t_2} \frac{\partial L}{\partial \dot{y}}\frac{d\delta y}{dt}dt = - \int_{t_1}^{t_2} \frac{d}{dt}\frac{\partial L}{\partial \dot{y}}\delta ydt$$

이다. 따라서

$$ \delta J = \int_{t_1}^{t_2}\left(\frac{\partial L}{\partial y} - \frac{d}{dt}\frac{\partial L}{\partial \dot{y}}\right)\delta y dt = 0$$

이 임의의 $\delta y$ 에 대하여 항상 성립하려면 괄호 안이 $0$이어야 한다.

즉,

$$ \frac{\partial L}{\partial y} - \frac{d}{dt}\frac{\partial L}{\partial \dot{y}} = 0 $$

이 물체가 따르는 경로 $y(t)$가 만족하는 식이다. 이 방정식을 **오일러-라그랑주 방정식** 또는 **라그랑주 방정식**이라 한다. 일반적인 좌표 $ q_i (i = 1, 2 \dots, n) $에 대하여 라그랑지안을

$$L=L(q_i, \dot{q_i}; t), \space\space\space\space (i=1, 2, \dots, n)$$

형태로 쓸 수 있다면, 각각의 $q_i$에 대하여

$$ \color{blue}{\frac{\partial L}{\partial q_i} - \frac{d}{dt}\frac{\partial L}{\partial \dot{q_i}} = 0} , \space\space\space\space (i=1, 2, \dots, n)$$

이 성립한다. (일반화된 좌표에 대한 라그랑주 방정식)

```python
import sympy as sym
import numpy as np
import scipy.optimize as so
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

## 10.3 예제

```python
# 독립 변수 및 상수
t, M, m, r, g= sym.symbols(r"t M m r g")

# X, theta는 t의 함수로 지정해 준다.
X, th = sym.symbols(r"X \theta", cls=sym.Function)

X = X(t)
th = th(t)
```

```python
sym.init_printing()
```

```python
X, th, M, m, r, g
```

```python
sym.diff(M, t)
```

```python
# 카테시안 좌표 (파생된 기호들):

x = X + r * sym.sin(th)
y = -r*sym.cos(th)
x, y
```

```python
xdot = sym.diff(x, t)
ydot = sym.diff(y, t)
Xdot = sym.diff(X, t)
xdot, ydot, Xdot
```

```python
T = sym.simplify(sym.expand((M/2) * Xdot**2 + (m/2) * (xdot**2 + ydot**2)))
V = m * g * y
T, V
```

$\Longrightarrow\space$ 교재 p.460 의 식 (10.3.4a, b):

$$ \begin{align}T &= \frac{M}{2}\dot{X}^2 + \frac{m}{2}\left[ \dot{X}^2 +(r\dot\theta)^2 + 2\dot X r \dot\theta \cos\theta \right] \\
\\
 V &= -m g r \cos\theta
\end{align}$$

와 비교해 보시오.

```python
# 라그랑지안으로부터 오일러-라그랑주 방정식을 구한다
L = T - V  # 라그랑지안

Xdot = sym.diff(X, t)
thdot = sym.diff(th, t)

eq1 = sym.diff(L, X) - sym.diff(sym.diff(L, Xdot), t) # X에 대한 라그랑주 방정식(의 좌변)
eq2 = sym.diff(L, th) - sym.diff(sym.diff(L, thdot), t) # theta에 대한 라그랑주 방정식(의 좌변)
eq1, eq2
```

라그랑주 방정식은 X와 theta의 t에 대한 2차 미분이 섞여 있는 방정식인데 미분 방정식의 수치 해를 구하기 위해서는 각 변수에 대한 2차 미분이 분리되어 있어야 한다. 다행히도 오일러-라그랑주 방정식들은 독립변수의 2차 미분들 ($ \dfrac{d^2X}{dt^2}$, $\dfrac{d^2\theta}{dt^2}$)에 대한 1차 연립 방정식이므로 쉽게 분리된다.


```python
Xddot = sym.diff(X,t,2)
thddot = sym.diff(th,t,2)

# 연립방정식을 풀어 각 변수에 대한 분리된 방정식을 구한다
sep = sym.solve([eq1, eq2], [Xddot, thddot])

Xdd = sep[Xddot]   # d²X/dt² 에 대한 식
thdd = sep[thddot] # d²θ/dt² 에 대한 식

Xdd, thdd
```

```python
sep
```

```python
gg = 9.8
MM = 2.0
mm = 1.0
rr = 1.0
subsrule = {g:gg, M:MM, m:mm, r:rr}

# 기호식 Xdd, thdd로부터 각각 계산 가능한 함수 Xddf, thddf를 생성한다
Xddf = sym.lambdify([t, X, th, Xdot, thdot], Xdd.subs(subsrule))
thddf = sym.lambdify([t, X, th, Xdot, thdot], thdd.subs(subsrule))
```

```python
def dSdt(t, S):
  """
  주어진 상태 (X, th, dX/dt, dth/dt)일 때 각 상태의 변화율을 구하는 함수
  """
  X, th, Xdot, thdot = S
  return np.array([Xdot, thdot, \
                   Xddf(t, X, th, Xdot, thdot), thddf(t, X, th, Xdot, thdot)])
```

```python
X0 = 0
th0 = np.pi/4
Xdot0 = 0
thdot0 = 0
tf=10
N=1000
teval=np.linspace(0, tf, N+1)

S0 = np.array([X0, th0, Xdot0, thdot0])

sol = solve_ivp(dSdt, (0, 10.0), S0, t_eval=teval, max_step=tf/N)
```

```python
sol
```

```python
ts = sol.t
Xs = sol.y[0]
ths = sol.y[1]
Xdots = sol.y[2]
thdots = sol.y[3]
```

```python
plt.grid()
plt.xlabel("t [s]")
plt.ylabel("X [m] / theta [rad]")
plt.plot(ts, Xs, label="X [m]")
plt.plot(ts, ths, label="ths [rad]")
plt.legend()
```

```python
xs =  Xs + rr * np.sin(ths)
ys = -rr * np.cos(ths)

plt.grid()
plt.xlabel("t [s]")
plt.ylabel("x, y [m]")
plt.plot(ts, xs, label="x [m]")
plt.plot(ts, ys, label="y [m]")
plt.legend()
```

위 그림을 보면 m은 좌우로(x축) 한 번 왕복하는 동안 상하로(y축) 두 번 오르내리는 것을 알 수 있다.

```python
O = np.zeros_like(Xs)
plt.axis("equal")
plt.plot(Xs, O, label="M")
plt.plot(xs, ys, label="m")
plt.legend()
```

```python
"""
M과 m을 연결하는 화살표를 겹쳐서 나타내 보면
두 물체의 위치가 연동되어 운동하는 모습을
더 잘 파악할 수 있다.
"""
skip=5
indf = 91
plt.axis("equal")
O = np.zeros_like(Xs)
plt.axis("equal")
plt.plot(Xs, O, label="M")
plt.plot(xs, ys, label="m")
plt.legend()
plt.quiver(Xs[:indf:skip], O[:indf:skip], (xs-Xs)[:indf:skip], ys[:indf:skip], \
           angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
```

```python
Xdot = sym.diff(X,t)
thdot = sym.diff(th, t)
Tf = sym.lambdify([X, th, Xdot, thdot], T.subs(subsrule))
Vf = sym.lambdify([X, th, Xdot, thdot], V.subs(subsrule))
```

```python
KE = Tf(Xs, ths, Xdots, thdots)
PE = Vf(Xs, ths, Xdots, thdots)
```

```python
ME = KE + PE
```

```python
plt.grid()
plt.xlabel("t [s]")
plt.ylabel("Energy [J]")
plt.plot(ts, KE, label="KE")
plt.plot(ts, PE, label="PE")
plt.plot(ts, ME, label="ME")
plt.legend()
```
