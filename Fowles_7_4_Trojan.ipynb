# Fowles Analytical Mechanics

```python
import sympy as smp
from sympy import symbols, Eq, Function
from sympy.solvers.ode.systems import dsolve_system
```

```python
import numpy as np
```

```python
from scipy.integrate import solve_ivp
```

```python
x, y, u, v = symbols("x y u v")
alpha = symbols(r"\alpha")
t = symbols("t")
```

```python
r1 = smp.sqrt((x - alpha)**2 + y**2)
r2 = smp.sqrt((x + 1 - alpha)**2 + y**2)
```

```python
r1
```

```python
r2
```

```python
Phi = -(1-alpha)/r1 + (-alpha)/r2 -(x**2 + y**2)/2
Phi
```

```python
xs = np.linspace(-1.5, 1.5, 201)
xm, ym = np.meshgrid(xs, xs)
```

```python
Phif = smp.lambdify([x, y, alpha], Phi)
```

```python
alpha_v = 0.000953875

Phis = Phif(xm, ym, alpha_v)
```

```python
dudt = -(1-alpha)*(x-alpha)/r1**3 - alpha*(x+1-alpha)/r2**3 + x + 2 * v
dvdt =  -(1-alpha)*y/r1**3 - alpha*y/r2**3 + y - 2 * u
```

```python
dudt
```

```python
dvdt
```

```python
dudtf = smp.lambdify([x, y, u, v, alpha], dudt)
dvdtf = smp.lambdify([x, y, u, v, alpha], dvdt)
```

```python
def dSdt(t, S, alpha, foo, bar):
  x, y, u, v = S

  return (u, v, dudtf(x,y,u,v,alpha), dvdtf(x,y,u,v,alpha))
```

```python
class Simulation():
  def __init__(self, fun, S0, alpha, tspan, ts, dt):
    self.dSdt = fun
    self.alpha = alpha
    self.S0 = S0
    self.tspan = tspan
    self.ts = ts
    self.dt = dt

  def run(self):
    return solve_ivp(self.dSdt, self.tspan, self.S0, 'Radau', self.ts, False,\
                     None, False, (self.alpha, 0, 0), max_step=self.dt)
```

```python
t0 = 0
tf = 300
N = 10000
dt = tf - t0/(10*N)
tspan = (t0, tf)
ts = np.linspace(t0, tf, N+1)
```

```python
S1 = (-0.509, 0.883, 0.0259, 0.0149)  # Orbit1
S2 = (-0.524, 0.909, 0.0647, 0.0367)  # Orbit2
S3 = (-0.524, 0.920, 0.0780, 0.0430)  # Orbit3
S4 = (-0.509, 0.883, -0.0259, -0.0149) # Orbit4
S5 = (-0.532, 0.920, 0.0780, 0.0430)  # Orbit5

Sim1 = Simulation(dSdt, S1, alpha_v, tspan, ts, dt)
Sim2 = Simulation(dSdt, S2, alpha_v, tspan, ts, dt)
Sim3 = Simulation(dSdt, S3, alpha_v, tspan, ts, dt)
Sim4 = Simulation(dSdt, S4, alpha_v, tspan, ts, dt)
Sim5 = Simulation(dSdt, S5, alpha_v, tspan, ts, dt)
```

```python
sol1 = Sim1.run()
sol2 = Sim2.run()
sol3 = Sim3.run()
sol4 = Sim4.run()
sol5 = Sim5.run()
```

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
ax.set_aspect(aspect='equal')

plt.grid()
plt.plot(sol1.y[0], sol1.y[1], label='Orbit 1')
plt.plot(sol2.y[0], sol2.y[1], label='Orbit 2')
plt.plot(sol3.y[0], sol3.y[1], label='Orbit 3')

plt.scatter(0,0, c='r', label='Sun')
plt.scatter(-1,0, c='g', label='Jupiter')

plt.legend(loc="lower right")
```
