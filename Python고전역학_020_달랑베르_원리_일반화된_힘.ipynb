## 10.8 D'Alembert's Principle: Generalized Forces

* 힘의 평형:
$$\sum_{i=1}^{N}\ \mathbf{F}_{i} = 0$$

--- $\mathbf{F}_i$는 $i$번째 입자에 작용하는 모든 힘


가상적인 변위 $\{\delta\mathbf{r}_i\}$에 동반되는 일:
$$\delta W = \sum_{i=1}^{N}\ \mathbf{F}_{i}\cdot \delta\mathbf{r}_i = 0$$

* 동력학적 상황:
$$\mathbf{F}_{i} = \dot{\mathbf{p}}_i ~~~~~~~~~~\{i=1,2,\dots,N\}$$

* 달랑베르 원리(D'Alembert's principle):
'관성력'인 $-\dot{\mathbf{p}}_i$를 힘의 일종처럼 취급하여, 가상적인 변위에 따른 가상적인 일의 원리에 편입시킨다:

$$\delta W = \sum_{i=1}^{N}\ (\mathbf{F}_{i} - \dot{\mathbf{p}}_i)\cdot \delta\mathbf{r}_i = 0$$

$\Longleftrightarrow$ 뉴턴의 운동 법칙

$$ \mathbf{F}_{i} - \dot{\mathbf{p}}_i = 0 ~~~~~~~~~~\{i=1,2,\dots,N\}$$

과 동치이지만, 동력학적 문제를 정력학적 형태로 환원함.

$\longrightarrow$ 이것을 바탕으로 라그랑주 방정식을 유도:

우선, 가상적인 일을 데카르트 좌표계에서 표현하면,
$$ \sum_{i=1}^{3}\ (F_i - \dot{p}_i)\delta x_i = 0 ~~~~~~~~~~~~~(10.8.6)$$
여기서 아래 첨자 $i$는 입자의 번호가 아니라 데카르트 좌표계의 좌표축 방향 중 하나를 나타낸다. $F_i$는 모든 입자에 작용하는 모든 힘의 $i$ 방향 성분의 합이다.

이것을 일반화된 좌표계로 옮겨서 표현하면, 각각의 $x_i$는 일반화된 좌표 $q_j$들의 함수, 즉
$$x_i = x_i(q_1, q_2, \dots, q_n)$$
라고 볼 수 있으므로,
$$\delta x_i = \dfrac{\partial x_i}{\partial q_1}\delta q_1 + \dfrac{\partial x_i}{\partial q_2}\delta q_2 + \cdots + \dfrac{\partial x_i}{\partial q_N}\delta q_N = \sum_j \dfrac{\partial x_i}{\partial q_j}\delta q_j$$
를 이용하여 (10.8.6)의 합의 기호 안 첫 번째 항에 적용하면,

$$ \delta W = \sum_i^3 F_i\ \delta x_i = \sum_j \left[\sum_i \left(F_i\dfrac{\partial x_i}{\partial q_j}\right)\right]\delta q_j = \sum_j Q_j \delta q_j$$

이며, 이 값은 0이 아니다. 각각의 $Q_j$들은
$$Q_j = \sum_i F_i \dfrac{\partial x_i}{\partial q_j}$$
로 주어지며, 이것을 각각의 일반화된 좌표 $q_j$에 대응되는 *$\color{blue}{\it{일반화된\ 힘}}$* (*generalized force*)이라 부른다.

식 (10.8.6)의 관성력 항을 계산하면,

$$
\begin{align}
 \sum_i \dot{p}_i \delta x_i &= \sum_i m \ddot{x}_i\ \delta x_i = \sum_j \left[\sum_i m\ddot{x}_i\ \dfrac{\partial x_i}{\partial q_j}\right]\ \delta q_j \\
 &=\sum_j \delta q_j\ \sum_i m \left[ \dfrac{d}{dt}\left( \dot{x}_i \dfrac{\partial x_i}{\partial q_j} \right) - \dot{x}_i\dfrac{d}{dt}\left(\dfrac{\partial x_i}{\partial q_j}\right)\right]
\end{align} ~~~~~~~(10.8.9)
$$

여기서 다시 $x_i = x_i(q_1, q_2, \dots, q_N)$인 관계로부터,
$$ \dot{x}_i = \sum_j \dfrac{\partial x_i}{\partial q_j} \dot{q}_j$$
및 이 식의 양변을 미분하여
$$\dfrac{\partial \dot{x}_i}{\partial \dot{q}_j} = \dfrac{\partial x_i}{\partial q_j}$$
를 (10.8.9)의 대괄호 안 첫번째 항에 대입하면

$$\sum_i m \dfrac{d}{dt}\left(\dot{x}_i \dfrac{\partial x_i}{\partial q_j} \right) = \sum_i m \dfrac{d}{dt}\left(\dot{x}_i \dfrac{\partial \dot{x}_i}{\partial \dot{q}_j} \right) = \dfrac{d}{dt}\left[  \dfrac{\partial}{\partial \dot{q}_j} \left(\sum_i \frac{1}{2}m\dot{x}_i^2 \right)\right] = \dfrac{d}{dt}\left(\dfrac{\partial T}{\partial \dot{q}_j} \right)$$

$q_j$와 $t$의 일반적인 함수의 시간 도함수는
$$\dfrac{d}{dt}f(q_j, t) = \sum_k \dfrac{\partial f}{\partial q_k}\dot{q}_k + \dfrac{\partial f}{\partial t}$$
로 주어지므로, 이것을 (10.8.9)의 두 번재 항에 적용하면

$$ \sum_i m \dot{x}_i \dfrac{d}{dt}\left(\dfrac{\partial x_i}{\partial q_j}\right) = \sum_i m \dot{x}_i \left[ \sum_k \dfrac{\partial^2 x_i}{\partial q_k \partial q_j}\dot{q}_k + \dfrac{\partial^2 x_i}{\partial q_j \partial t} \right] = \sum_i m \dot{x}_i \dfrac{\partial \dot{x}_i}{\partial q_j} = \dfrac{\partial}{\partial q_j} \left[ \sum_i \dfrac{1}{2}m \dot{x}_i^2\right] = \dfrac{\partial T}{\partial q_j}$$

따라서 관성력 항 전체는
$$ \sum_i \dot{p}_i\ \delta x_i = \sum_i \left[ \dfrac{d}{dt}\left(\dfrac{\partial T}{\partial \dot{q}_j}\right) - \dfrac{\partial T}{\partial q_j}\right] \delta q_j$$

따라서 달랑베르 원리를 일반화 좌표계에서 쓰면,
$$ \sum_i (F_i - \dot{p}_i)\ \delta x_i = \sum_j \left\{Q_j - \left[ \dfrac{d}{dt}\left(\dfrac{\partial T}{\partial \dot{q}_j}\right) - \dfrac{\partial T}{\partial q_j}\right] \right\}\delta q_j = 0 ~~~~~~~~(10.8.15)$$

우선 구속력이 없다고 가정하면 각각의 $q_j$를 독립적으로 변화시킬 수 있다고 할 수 있으므로, 대괄호 안의 각각의 항이 $0$이 되어야 하고, 따라서
$$\dfrac{d}{dt}\left(\dfrac{\partial T}{\partial \dot{q}_j}\right) - \dfrac{\partial T}{\partial q_j} = Q_j$$
를 얻는다.

일반적인 힘 $Q_j$ 가 보존력이라면 위치에만 의존하는 퍼텐셜 $V(x_i)$에 대하여
$$Q_j = -\dfrac{\partial V}{\partial q_j}$$
를 이용하여
$$\dfrac{d}{dt}\left(\dfrac{\partial T}{\partial \dot{q}_j}\right) - \dfrac{\partial (T-V)}{\partial q_j} = 0$$
라 할 수 있고,

또는 일반적인 힘이 속도에도 의존하는 $V=V(q_j, \dot{q}_j, t)$로부터 유도된다고 하면
$$ Q_j = \dfrac{d}{dt}\left(\dfrac{\partial V}{\partial \dot{q}_j}\right) - \dfrac{\partial V}{\partial q_j} $$를 이용하여
$$\dfrac{d}{dt}\left[\dfrac{\partial (T-V)}{\partial \dot{q}_j}\right] - \dfrac{\partial (T-V)}{\partial q_j} = 0,$$
또는
$$\dfrac{d}{dt}\left(\dfrac{\partial L}{\partial \dot{q}_j}\right) - \dfrac{\partial L}{\partial q_j} = 0$$
이라 쓸 수 있다. (여기서 라그랑지안의 정의 $L=T-V$를 이용함.)

여기서 $Q_j$가 비보존력을 포함하는 일반적인 힘이라면, 퍼텐셜 에너지로 정의되는 부분만 라그랑지안 안에 편입시키고, 나머지는 $Q_j^\prime$으로 써서
$$\dfrac{d}{dt}\left[\dfrac{\partial L}{\partial \dot{q}_j}\right] - \dfrac{\partial L}{\partial q_j} = Q_j^\prime ~~~~~~~\{j=1,2,\dots, n\} ~~~ (10.8.21)$$
라고 쓸 수 있다.

여기서 $Q_j'$는 $\dfrac{\partial L}{\partial q_j}$, 또는 $\dfrac{L}{q_j}$와 같은 차원을 갖는다.



## 레일리 소산 함수 (Rayleigh's dissipation function)

* 참고: Goldstein, et al., `Classical Mechanics', pp.22-24


비보존력 $Q_j'$가 특히 속도에 의존하는 힘이고, 다음과 같은 형태
$$ F_{f,x} = -c_x v_x$$
로 주어질 때, 레일리 소산 함수(Rayleigh's dissipation function)

$$ R = \dfrac{1}{2} \sum_i \left( c_x v_{ix}^2 + c_y v_{iy}^2 + c_z v_{iz}^2 \right) = \dfrac{1}{2} \sum_i \sum_{k=1}^3 c_j v_{i,k}^2$$

를 정의한다. 여기서 $i$는 계를 이루는 각각의 입자를 나타내고, $k$는 3차원 데카르트 좌표계의 세 좌표 방향 성분을 나타낸다. 여기서

$$ F_{fx_i} = - \dfrac{\partial R}{\partial v_{x_i}} $$

이고, 일반화 좌표계에서 이렇게 정의된 힘의 성분은

$$
\begin{align}
Q_j &= \sum_i \mathbf{F}_{f,i} \cdot \dfrac{\partial \mathbf{r}_i}{\partial q_j} = - \sum_i \sum_{k=1}^3 \dfrac{\partial R}{\partial v_{i,k}} \dfrac{\partial x_{i,k}}{\partial q_j}\\
 &= - \sum_i \sum_{k=1}^3 \dfrac{\partial R}{\partial v_{i,k}} \dfrac{\partial \dot{x}_{i,k}}{\partial \dot{q}_j} \\
 &= -\dfrac{\partial R}{\partial \dot{q}_j}
\end{align}
$$

이 표현을 이용하면 속도에 의존하는 저항력이 작용하는 경우 오릴러-라그랑주 방정식은

$$\dfrac{d}{dt}\left[\dfrac{\partial L}{\partial \dot{q}_j}\right] - \dfrac{\partial L}{\partial q_j} + \dfrac{\partial R}{\partial \dot{q}_j} = 0 $$

으로 쓸 수 있다.

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import scipy.fft as fft
```

```python
sym.init_printing()
```

```python
'''
기본적인 변수 및 상수들의 기호 정의:
(저항 계수 k 추가.)
'''

theta, phi = sym.symbols(r'\theta \phi', cls=sym.Function) #
g, m, l, c1, t = sym.symbols("g m l c_1 t", positive=True)
```

```python
'''
일반화 좌표 theta 와 phi의 시간 미분꼴을 구하기 위해 시간의 함수로 정의해준다.
'''
theta = theta(t)
phi = phi(t)

'''
theta, phi의 시간(t)에 대한 1차 및 2차 미분을 정의:
'''

theta_d = sym.diff(theta, t)
phi_d = sym.diff(phi,t)

theta_dd = sym.diff(theta_d, t)
phi_dd = sym.diff(phi_d,t)
```

$x = l \sin\theta \cos\phi$

$y = l \sin\theta \sin\phi$

$z = -l \cos\theta$

```python
'''
위치벡터의 카테시안 좌표계 성분:
'''
x =  l * sym.sin(theta) * sym.cos(phi)
y =  l * sym.sin(theta) * sym.sin(phi)
z = -l * sym.cos(theta)

'''
위치값을 계산하는 함수 생성
'''
xf = sym.lambdify((l, theta, phi), x)
yf = sym.lambdify((l, theta, phi), y)
zf = sym.lambdify((l, theta, phi), z)
```

$v_x = \dfrac{dx}{dt}=\dfrac{\partial x}{\partial \theta}\dfrac{d\theta}{dt}+\dfrac{\partial x}{\partial \phi}\dfrac{d\phi}{dt}=l\cos\theta\cos\phi\dfrac{d\theta}{dt}-l\sin\theta\sin\phi\dfrac{d\phi}{dt}$

$v_y = \dfrac{dy}{dt}=\dfrac{\partial y}{\partial \theta}\dfrac{d\theta}{dt}+\dfrac{\partial y}{\partial \phi}\dfrac{d\phi}{dt}=l\cos\theta\sin\phi\dfrac{d\theta}{dt}+l\sin\theta\cos\phi\dfrac{d\phi}{dt}$

$v_z = \dfrac{dz}{dt}=\dfrac{\partial z}{\partial \theta}\dfrac{d\theta}{dt}=l\sin\theta\dfrac{d\theta}{dt}$

```python
'''
속도 성분들의 식
'''
vx = sym.diff(x, t)
vy = sym.diff(y, t)
vz = sym.diff(z, t)
```

```python
vx
```

```python
vy
```

```python
vz
```

$ T = \dfrac{1}{2} m \left( v_x^2 + v_y^2 +v_z^2 \right) $

$ V = m g z $

$ L = T - V $

레일리 소산 함수를 아래와 같이 정의한다: (속도에 비례하는 저항력)

$$ R = \dfrac{1}{2} c_1 \left( v_x^2 + v_y^2 +v_z^2 \right) $$

```python
'''
라그랑지안 생성:
'''
T = (1/2) * m * (vx**2 + vy**2 + vz**2);
V = m * g * z;
L = T - V;
L = L.simplify();
L
```

```python
'''
레일리 소산 함수
'''
R = (1/2) * c1 * (vx**2 + vy**2 + vz**2)
R
```

```python
'''
각각 theta, phi에 대한 라그랑주 방정식 eqt, eqp를 구한다
'''
eqt = sym.diff(sym.diff(L,theta_d), t) - sym.diff(L,theta) + sym.diff(R, theta_d)
eqp = sym.diff(sym.diff(L,phi_d), t) - sym.diff(L,phi) + sym.diff(R, phi_d)
```

```python
eqt
```

```python
eqp
```

```python
phi_dd
```

```python
'''
theta, phi 각각의 시간에 대한 이차미분 theta_dd, phi_dd 에 대하여 정리해준다:
'''
sols = sym.solve([eqt, eqp], [theta_dd, phi_dd])
```

```python
sols
```

```python
sols[theta_dd]
```

```python
sols[phi_dd]
```

```python
'''
시뮬레이션에 사용할 상수들:
'''
g_val = 9.8
m_val = 1
l_val = 0.284
c1_val = 0.2

params = {g:g_val, m:m_val, l:l_val, c1:c1_val}
```

```python
'''
계산 가능한 함수 생성
'''
theta_dd_f = sym.lambdify((theta,phi,theta_d,phi_d), sols[theta_dd].subs(params))
phi_dd_f = sym.lambdify((theta,phi,theta_d,phi_d), sols[phi_dd].subs(params))
```

```python
def dSdt(t, S):
  """
  현재 상태 S=[theta, phi, dthdt, dphdt] 를 받아
  변화율 dSdt를 계산하여 반환한다
  """
  theta, phi, thetad, phid = S
  return [
      thetad,
      phid,
      theta_dd_f(theta,phi,thetad,phid),
      phi_dd_f(theta,phi,thetad,phid)
  ]
```

```python
ivals = [sym.pi/4, 0, 0, 10.57] # [theta0, phi0, dthetadt0, dphidt0]
```

```python
'''
이차 방정식의 수치해를 구한다:
'''
tf = 200
N = 5000
dt = tf / (N - 1)

teval = np.linspace(0, tf, N*5)
ans = solve_ivp(dSdt, (0, tf), ivals, t_eval=teval, max_step=0.01)
```

```python
ans
```

```python
'''
시뮬레이션 결과를 각 변수들에 대하여 분리해 준다:
'''
ts = ans.t
th = ans.y[0]
ph = ans.y[1]
dthdt = ans.y[2]
dphdt = ans.y[3]
```

```python
plt.grid()
plt.ylim([0,np.pi/2])
plt.xlabel(r'$t(s)$')
plt.ylabel(r'$\theta(rad)$')
plt.plot(ts[:2500], th[:2500], c='dodgerblue', label='Linear Drag')
plt.legend()
```

```python
def getpos(l, theta, phi):
  return [
      xf(l, theta, phi),
      yf(l, theta, phi),
      zf(l, theta, phi)
  ]
```

```python
xs, ys, zs = getpos(l_val, th, ph)
```

```python
'''
진자의 운동을 xy 평면에 투영하여 그려본다:
'''

plt.axes().set_aspect(1)
plt.xlabel("x(m)")
plt.ylabel("y(m)")
plt.grid()
plt.plot(xs[:2500], ys[:2500], c='dodgerblue', label='Linear Drag')
plt.plot(xs[0], ys[0], 'ro') # starting point
plt.annotate(r'Start', (xs[0], ys[0]))
plt.legend()
```

## 속력의 제곱에 비례하는 저항력
속력의 제곱에 비례하는 저항력은
$$F_x = -c_2 v_x |\mathbf{v}| = -c_2 v_x \left(v_x^2 + v_y^2 + v_z^2\right)^{1/2}$$
와 같이 주어지므로, 레일리 소산 함수를
$$R = \dfrac{c_2}{3} \left(v_x^2 + v_y^2 + v_z^2\right)^{3/2}$$
로 정의할 수 있다.

* $\longrightarrow$ $F_i = -\dfrac{\partial R}{\partial v_i} = -c_2 v_i \left|\mathbf{v}\right|$임을 확인해 보시오.

```python
c2 = sym.symbols('c_2')
c2
```

```python
'''
레일리 소산 함수
'''
R2 = (c2 / 3) * (vx**2 + vy**2 + vz**2) ** (3/2)
R2
```

```python
'''
각각 theta, phi에 대한 라그랑주 방정식 eqt2, eqp2를 구한다
'''
eqt2 = sym.diff(sym.diff(L,theta_d), t) - sym.diff(L,theta) + sym.diff(R2, theta_d)
eqp2 = sym.diff(sym.diff(L,phi_d), t) - sym.diff(L,phi) + sym.diff(R2, phi_d)
```

```python
sols2 = sym.solve([eqt2, eqp2], [theta_dd, phi_dd])
```

```python
'''
시뮬레이션 파라미터
'''
c2_val = 0.2

params2 = {g:g_val, m:m_val, l:l_val, c2:c2_val}
```

```python
'''
계산 가능한 함수 생성
'''
theta_dd_f2 = sym.lambdify((theta,phi,theta_d,phi_d), sols2[theta_dd].subs(params2))
phi_dd_f2 = sym.lambdify((theta,phi,theta_d,phi_d), sols2[phi_dd].subs(params2))
```

```python
def dSdt2(t, S):
  """
  현재 상태 S=[theta, phi, dthdt, dphdt] 를 받아
  변화율 dSdt를 계산하여 반환한다
  """
  theta, phi, thetad, phid = S
  return [
      thetad,
      phid,
      theta_dd_f2(theta,phi,thetad,phid),
      phi_dd_f2(theta,phi,thetad,phid)
  ]
```

```python
ivals = [sym.pi/4, 0, 0, 10.57] # [theta0, phi0, dthetadt0, dphidt0]
```

```python
'''
이차 방정식의 수치해를 구한다:
'''
tf = 200
N = 5000
dt = tf / (N - 1)

teval = np.linspace(0, tf, N*5)
ans2 = solve_ivp(dSdt2, (0, tf), ivals, t_eval=teval, max_step=0.01)
```

```python
'''
시뮬레이션 결과를 각 변수들에 대하여 분리해 준다:
'''
ts2 = ans2.t
th2 = ans2.y[0]
ph2 = ans2.y[1]
dthdt2 = ans2.y[2]
dphdt2 = ans2.y[3]
```

```python
plt.grid()
plt.ylim([0,np.pi/2])
plt.plot(ts[:2500], th[:2500], c='dodgerblue', label='Linear Drag')
plt.plot(ts2[:2500], th2[:2500], c='orangered', label='Quadratic Drag')
plt.legend()

# 색상 이름 참고: https://matplotlib.org/stable/gallery/color/named_colors.html
```
